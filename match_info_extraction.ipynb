## T20 Info Table

```python
cursor = conn.cursor()
create_table_query = f"""
    CREATE TABLE IF NOT EXISTS t20matchinfo (
        match_id INT,
        file_id INT PRIMARY KEY,
        balls_per_over INT,
        dates_0 DATE,
        event_name VARCHAR(255),
        gender VARCHAR(10),
        match_type VARCHAR(55),
        season VARCHAR(10),
        team_type VARCHAR(55),
        teams_0 VARCHAR(55),
        teams_1 VARCHAR(55),
        toss_winner VARCHAR(55),
        toss_decision VARCHAR(55),
        winner VARCHAR(55),
        by_wickets DECIMAL(2,0),
        by_runs DECIMAL(3,0),
        outcome VARCHAR(55)
        );
        """

    # Execute the query
cursor.execute(create_table_query)

    # Commit the changes
conn.commit()


print(f"Table created successfully.")
```
