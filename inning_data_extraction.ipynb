```python
import pandas as pd
from flatten_json import flatten
import os, sys
import json
```

## INNING DATA EXTRACTION FUNCTION

```python
def inning_info(data, files):
    inning_df = pd.DataFrame()
    for n in range (len(data['innings'])):
        for i in range(len(data['innings'][n]['overs'])):
            for j in range(len(data['innings'][n]['overs'][i]['deliveries'])):
                delivery_details = [data['innings'][n]['overs'][i]['deliveries'][j]]
                flat_list = [flatten(item) for item in delivery_details]
                flat_list = pd.DataFrame(flat_list)
                flat_list['file_id'] = files
                flat_list['inning']=n
                flat_list['Over'] = i+1
                flat_list['delivery'] = j+1
                flat_list['team'] = data['innings'][n]['team']
                #flat_list['event'] = data['info']['event']['name']
                flat_list['season'] = data['info']['season']
                flat_list['team_type'] = data['info']['team_type']
                flat_list['gender'] = data['info']['gender']
                inning_df = pd.concat([inning_df, flat_list], axis=0)
    return inning_df
```

## T20 INNING DATA

```python
#Extracting T20 match info
path = "D:\\GUVI\\visual studio\\cric\\cric_unzip_files\\t20_matches"
dirs=os.listdir(path)
match_id = 0
file_names = [file for file in dirs if 'json' in file]
t20inningdata = pd.DataFrame()
for files in file_names:
    f = open(f'D:\\GUVI\\visual studio\\cric\\cric_unzip_files\\t20_matches\\{files}')
    info_data=json.load(f)
    print('extracting', files)
    match_id = match_id+1
    inninginfo_df = inning_info(info_data, files)
    inninginfo_df.insert(0, 'match_id', match_id)
    t20inningdata = pd.concat([t20inningdata, inninginfo_df], ignore_index=True)
    print(f'extracted {files}')
t20inningdata
```

```python
pattern = '.json'
t20inningdata['file_id'] = t20inningdata['file_id'].str.replace(pattern, '', regex = True)
t20inningdata['file_id'] = t20inningdata['file_id'].astype('int')
```
